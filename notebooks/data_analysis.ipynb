```python
import os
from dare2d.datamodule.generator._2d.regress_2dataset import Regress2Dataset
import numpy as np
from tqdm import tqdm
from rich.pretty import pprint
import matplotlib.pyplot as plt

target_crop_size = 64
data_dir = "../data/2d"

def create_histogram(length_values_array, angle_values_array, names, fig_affix=""):
    fig, axs = plt.subplots(2, len(names), figsize=(15,15), sharey="row")
    
    length_stats = {}
    angle_stats = {}
    sets_size = {}
    
    nbins = list(range(20, 61))
    abins_n = 40
    abins = np.linspace(0, 180, abins_n)
    
    for i, name in enumerate(names):
        counts, bins = np.histogram(length_values_array[i], nbins)
        axs[0, i].bar(bins[:-1], counts, width=0.95, linewidth=1, edgecolor="black")
        axs[0, i].set_title(f"{name} length")
        # axs[0, i].set_ylim((0, 30))
        length_stats[name] = {}
        length_stats[name]["mean"] = np.mean(length_values_array[i])
        length_stats[name]["std"] = np.std(length_values_array[i])
        sets_size[name] = {}
        sets_size[name]["sample_size"] = len(length_values_array[i])

        counts, bins = np.histogram(angle_values_array[i], abins)
        width = abins[-1] // abins_n
        axs[1, i].bar(bins[:-1], counts, width=width, linewidth=1, edgecolor="black")
        axs[1, i].set_title(f"{name} angle")
        
        angle_stats[name] = {}
        angle_stats[name]["mean"] = np.mean(angle_values_array[i])
        angle_stats[name]["std"] = np.std(angle_values_array[i])

    plt.savefig(f"sets_statistics_{fig_affix}.png", dpi=300)
    plt.clf()
    
    print("Length statistics:")
    pprint(length_stats)
    print("Angle statistics:")
    pprint(angle_stats)
    print("Sample sizes statistics:")
    pprint(sets_size)

def compute_length_angle_stats(generator, name):
    angle_values = []
    length_values = []
    for i in tqdm(range(len(generator))):
        x, y = generator[i]
        cosin = y["angle_output"]
        length = y["length_output"]
        
        # Denormalize
        angle = np.arctan2(cosin[0], cosin[1])
        angle = angle / 2
        angle = angle * 180 / np.pi
        angle = angle % 180
        
        length = length * x.shape[1]
        length_values.append(length)
        angle_values.append(angle)
    
    return length_values, angle_values, name

length_values, angle_values, names = [], [], []
nb_sets = 8

def process_sets(base_dir, nb_sets, fig_affix):
    for i in range(nb_sets):
        generator_data_dir = os.path.join(base_dir, f"set_{i+1}")
        generator = Regress2Dataset(
            generator_data_dir,
            [-1, 0, 1],
            scale=[1.0, 1.0],
            crop_size=target_crop_size
        )
        print(len(generator))
        lengths, angles, name = compute_length_angle_stats(generator, f"set{i+1}")
        length_values.append(lengths)
        angle_values.append(angles)
        names.append(name)

    create_histogram(length_values, angle_values, names, fig_affix)

process_sets(data_dir, nb_sets, "original")
# plt.cla()
# process_sets(base_dir_corrected, nb_sets, "corrected")
plt.show()
```
